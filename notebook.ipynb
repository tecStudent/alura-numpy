# Segundo módulo - exploração dos dados

- Separando as informações em arrays
- Criando uma vizualização
- Tratando valores Nan

```python
# separando as datas
datas = dados_transposto[:,0]
```

```python
# separando os preços 
precos = dados_transposto[:,1:6]
```

```python
# importando matplotlib para vizualizar os dados
import matplotlib.pyplot as plt
```

```python
plt.plot(datas,precos[:,0])
```

```python
# gerando um range de números para conseguir vizualizar 
datas = np.arange(1, 88)
```

```python
plt.plot(datas,precos[:,0])
```

```python
# passando um nome para os preços
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

```python
Moscow
```

```python
# Separando os valores por anos
Moscow_ano1 = Moscow[0:12] 
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
```

```python
# Mostra um grafico em relação com o ano um
plt.plot(np.arange(1,13,1), Moscow_ano1)
```

```python
# Mostra um grafico mostrando todos os anos
plt.plot(np.arange(1,13,1), Moscow_ano1)
plt.plot(np.arange(1,13,1), Moscow_ano2)
plt.plot(np.arange(1,13,1), Moscow_ano3)
plt.plot(np.arange(1,13,1), Moscow_ano4)
# adicona uma legenda
plt.legend(['ano1', 'ano2', 'ano3', 'ano4'])
```

```python
# A função serve para comparar se os arrays são iguais 
np.array_equal(Moscow_ano3, Moscow_ano4)
```

```python
# Para saber se a distância é menor que 0.01
np.allclose(Moscow_ano3, Moscow_ano4,0.01)
```

```python
np.allclose(Moscow_ano3, Moscow_ano4,10)
```
